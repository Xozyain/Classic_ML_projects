#### Probability Calibration

Now we move on to assessing the quality of the probabilities generated by the algorithms. Let's start with calibration curves.

Let's say the algorithm returns some numbers from zero to one. Are they good at assessing probability? To do this, we divide the segment $[0, 1]$ into several small segments of the same length.

Let's consider the $i$-th segment with boundaries $[a_i, b_i]$ and the predictions $p_1, p_2, \dots, p_k$ that fall into it. Let the true answers $y_1, y_2, \dots, y_k$ correspond to them. If the algorithm produces the correct probabilities, then there should be approximately $(a_i + b_i) / 2$ units among these true answers. In other words, if you draw a curve in which the centers of the segments are plotted along the X axis, and the proportions of these single answers in the segments along the Y axis, then it should turn out to be diagonal.

```python
def plot_calibration_curve(y_test, preds, title, label='classifier'):
    bin_middle_points = []
    bin_real_ratios = []
    n_bins = 10
    for i in range(n_bins):
        l = 1.0 / n_bins * i
        r = 1.0 / n_bins * (i + 1)
        bin_middle_points.append((l + r) / 2) 
        bin_real_ratios.append(np.mean(y_test[(preds >= l) & (preds < r)] == 1)) 
    plt.figure(figsize=(6,6))
    plt.plot(bin_middle_points, bin_real_ratios)
    plt.plot([0, 1], [0, 1], label='Perfect', linestyle='--', color='green')
    plt.title(title)
    plt.legend()
    plt.ylim([-0.05, 1.05])
    plt.grid()
```

# Part 2. Processing Categorical Variables

As we know, recoding a categorical variable into a list of numbers (for example, 1, 2, 3, ..., n) is bad, because this would set some order on the set of its values that does not make sense.

In this part, we will look at two main ways to handle categorical values:
- One-hot coding
- Counters (CTR, mean-target coding, ...) — each categorical feature is replaced by the average value of the target variable for all objects that have the same value in this feature.

Let's start with one-hot coding. Suppose our categorical feature $f_j(x)$ takes values from the set $C=\{c_1, \dots, c_m\}$. Let's replace it with $m$ binary features $b_1(x), \dots, b_m(x)$, each of which is an indicator of one of the possible categorical values:
$$
b_i(x) = [f_j(x) = c_i]
$$

####  Mean-target Encoding

> The problem of feature growth can be avoided in another way of encoding categorical features - mean-target encoding (for simplicity, we will call this __counters__). Let's compare the effectiveness of the methods within the framework of our marketing task.

> The main idea is that it is not the categories themselves that are important, but the values of the target variable that the objects of this category have. We will replace each categorical feature with the average value of the target variable for all objects of the same category:
$$
g_j(x, X) = \frac{\sum_{i=1}^{\ell} [f_j(x) = f_j(x_i)][y_i = +1]}{\sum_{i=1}^{ \ell} [f_j(x) = f_j(x_i)]}
$$

Let's encode categorical variables using counters (exactly as described above, without any tricks). Let's train logistic regression and look at the quality of the model on the test set.

Let's compare the training time with the previous experiments (taking into account feature encoding).

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class MeanTargetEncoding(BaseEstimator, TransformerMixin):
    def __init__(self, column_names):
        self.column_names = column_names
        self.features = {}
        pass
    
    def fit(self, X, y):
        df = pd.concat([X, y], axis=1)
        for column in self.column_names:
            self.features[column] = {}
            for line in df[column].unique():
                self.features[column][line] = df.loc[df[column] == line][y.name].mean()
        return self
    
    def transform(self, X, y=None):
        df = X.copy()
        for column in self.column_names:
            for line in df[column].unique():
                df.loc[df[column] == line, column] = self.features[column].get(line)
        return df
```

```python
transf = ColumnTransformer(
    transformers=[('meanEncoder', MeanTargetEncoding(categ), categ), ('scaler', StandardScaler(), num)])

pipe = Pipeline(steps=[('transf', transf), ('logReg', LogisticRegression(max_iter=1000))])
%time pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.decision_function(X_test)
```

```python
PrecisionRecallDisplay.from_predictions(y_test, y_pred)
precision, recall, _ = precision_recall_curve(y_test, y_pred)
print('AUC-PR: %.5f' % auc(recall, precision))
```

__Methods to combat counter retraining__


Note that mean-target encoding features are classifiers themselves, and by learning from them, we allow the target variable to "leak" into features. This leads to __overfitting__, so it is necessary to count such features in such a way that when calculating for a particular object, its __target label is not used__.

This can be done in the following ways:
1. Calculate the counter value for all objects located higher in the dataset (for example, if we have a selection sorted by time).
2. Calculate by folds, that is, divide the sample into a certain number of parts and calculate the value of features for all folds except the current one (as is done in cross-validation).
3. Introduction of some noise into the calculated features.

#### Let's make some noise

We implement the correct calculation of counters in the simplest way - adding noise to the values. At the same time, try to find a balance between fighting overfitting and maintaining the usefulness of features. Let's train logistic regression again, evaluate the quality.

```python
class MeanTargetEncodingNoise(BaseEstimator, TransformerMixin):
    def __init__(self, column_names):
        self.column_names = column_names
        self.features = {}
        pass
    
    def fit(self, X, y):
        df = pd.concat([X, y], axis=1)
        for column in self.column_names:
            self.features[column] = {}
            for line in df[column].unique():
                self.features[column][line] = df.loc[df[column] == line][y.name].mean() + np.random.normal(0, 0.01)
        return self
    
    def transform(self, X, y=None):
        df = X.copy()
        for column in self.column_names:
            for line in df[column].unique():
                df.loc[df[column] == line, column] = self.features[column].get(line)
        return df
```

```python
transf = ColumnTransformer(
    transformers=[('meanEncoder', MeanTargetEncodingNoise(categ), categ), ('scaler', StandardScaler(), num)])

pipe = Pipeline(steps=[('transf', transf), ('logReg', LogisticRegression(max_iter=1000))])
%time pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.decision_function(X_test)
PrecisionRecallDisplay.from_predictions(y_test, y_pred)
precision, recall, _ = precision_recall_curve(y_test, y_pred)
print('AUC-PR: %.5f' % auc(recall, precision))
```

#### Conclusion

+ Quality on the test has deteriorated, something has gone wrong. Sometimes in machine learning there is magic that we cannot comprehend

> Note.The complexity of counter methods does not make model results guaranteed to be better. Especially since logistic regression isn’t such a complicated model to retrain. So we didn’t have to get better and better at every step.

As we could see, meters are a competitive alternative to one-hot encoding.

#### Built-in methods

Let’s say we want to leave only 40 of the best features. Let’s try to do this in several ways.

Let’s start with feature selection using a linear model. As you know, the weights of the linear model mean the contribution of each attribute to the prediction of the model, so the module of this contribution can be interpreted as the importance of the traits. This selection method is called embedded or embedded method, because it is built-in in.

Let’s leave 40 features with the largest module corresponding to the parameter of the linear model. Let’s teach the model again and evaluate its quality.

```python
from sklearn.linear_model import Lasso, LinearRegression
```

```python
ohe = OneHotEncoder()
scaler = StandardScaler()

x_ohe = ohe.fit_transform(X_train[cat_columns])
x_prep = scaller.fit_transform(x_ohe.toarray())

x_ohe_test = ohe.transform(X_test[cat_columns])
x_prep_test = scaler.fit_transform(x_ohe_test.toarray())

lin_reg = LinearRegression()
lin_reg.fit(x_prep, y_train)

# Selecting top 40 columns
columns = ohe.get_feature_names_out()
coef = lin_reg.coef_
features = pd.DataFrame(coef, columns, columns=['importance'])
top_features = features.sort_values(['importance'], ascending=False).head(40).index
df_top_train = pd.DataFrame(x_prep, columns=columns)[top_features]
df_top_test = pd.DataFrame(x_prep_test, columns=columns)[top_features]
```

```python
# Training LogReg on top 40 features
log_reg = LogisticRegression(max_iter = 100000, random_state=42)
log_reg.fit(df_top_train, y_train)
```

```python
y_pred = log_reg.predict(df_top_test)
precision, recall, thresholds = precision_recall_curve(y_test, y_pred)
print('AUC-PR = %.5f' % auc(recall, precision))
```

**AUC-PR score has decreased a little bit**

####  Just to know

There is a smarter way to select features, namely by calculating some function for each feature. Based on the values of this function, you can leave the most important features. The methods of this family are called filter methods.

The t-statistic is considered to be such a function:

$$t(j) = \frac{|\mu_+ - \mu_-|}{\sqrt{\frac{n_+ \sigma^2_+ + n_- \sigma^2_-}{n_+ + n_-} }},$$

where $\mu$, $\sigma$, $n$ are the mean, standard deviation, and number of objects of each class, respectively.

#### Key metric

We have already described an approximate economic model of the problem above. What is more important for our business - good precision or recall models?

__Answer:__ Recall is more important for our model. The profit from the call is evenly distributed over the segment [0; 20], and the cost of the operator is $1, so the chance to profit from the call is greater



> Recall that in fact, logistic regression predicts the probabilities of a positive class for an object. Perhaps by adjusting the __binarization threshold__ of these probabilities, we can get some kind of gain?

Let's test your reasoning above by adjusting the binarization threshold on cross-validation to maximize profits. Use a grid from 0 to 1 with a step of 0.01. Lowering the threshold gives us higher recall and lower precision, and vice versa.

```python
from tqdm import tqdm

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=500)

best_c = None
best_profit = 0
best_threshold = 0

mean_profits = []
std_profits = []

for threshold in tqdm(np.linspace(0, 1, 100)):
    profits = np.array([])
    earnings = np.array([])
    total_calls = 0
    
    for train_index, test_index in kf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y[train_index], y[test_index]


        column_transformer = ColumnTransformer([
            ('ohe', OneHotEncoder(handle_unknown='ignore'), categorical),
            ('scaling', StandardScaler(), numeric)
        ])

        pipeline = Pipeline(steps=[
            ('scaling', column_transformer),
            ('logreg', LogisticRegression(max_iter=1000, C=c))
        ])

        pipeline.fit(X_train, y_train)
        y_pred = pipeline.predict_proba(X_test)[:, 1]
        y_pred = y_pred > threshold

        random_earning = np.random.uniform(0, 20)

        earnings = np.append(earnings, (y_pred * y_test).sum() * random_earning)
        total_calls = y_pred.sum()
        profits = np.append(profits, (y_pred * y_test).sum() * random_earning - total_calls)
        
    mean_profit = profits.mean()
    
    mean_profits.append(mean_profit)
    std_profits.append(profits.std())
    
    if mean_profit > best_profit:
        best_c = c
        best_profit = mean_profit
        best_threshold = threshold
```

```python
print('Best threshold: ', best_threshold)
print('Best mean profit: %.1f' % best_profit)
```

**As we expected, the threshold is low, which tells us that the model is trying to increase the value of recall**

```python
fig = plt.figure(figsize=(16, 9))
plt.title('Threshold dependence')
plt.xlabel('threshold')
plt.xlabel('mean_profits')
plt.errorbar([i for i in range(len(mean_profits))], mean_profits, yerr=std_profits, c='green', ecolor='red')
plt.show()
```
